```python
data["Income"]=data["Income"].fillna(data["Income"].median())
```

```python
#feature engineering
#means create new feature from exisiting or removing unwanted feature form dataset
data.columns
```

```python
data["age"]=2026-data["Year_Birth"]
```

```python
data.columns
```

```python

#customer joining days
data["Dt_Customer"]=pd.to_datetime(data["Dt_Customer"], dayfirst=True)
reference_date=data["Dt_Customer"].max()
data["Customer_Tenure_Days"]=(reference_date-data["Dt_Customer"]).dt.days

# create new total spending by each customer on platforms 
data["Total_spending"]=data["MntFruits"]+data["MntGoldProds"]+data["MntWines"]+data["MntSweetProducts"]+data["MntMeatProducts"]+data["MntFishProducts"]
```

```python
data.columns
```

```python
data["Total_child"]=data["Kidhome"] +data["Teenhome"]
```

```python
data["Education"].value_counts()
```

```python
data["Education"]=data["Education"].replace({
    "Basic":"Undergraduate","2n Cycle":"Undergraduate",
    "Master":"Postgraduate","PhD":"Postgraduate"
})
```

```python
data["Education"].value_counts()
```

```python
#meritals status
data["Marital_Status"].value_counts()
```

```python
data["Living_with"]=data["Marital_Status"].replace({
    "Married":"Partner","Together":"Partner","Widow":"Alone",
    "Single":"Alone","Divorced":"Alone","Alone":"Alone","Absurd":"Alone","YOLO":"Alone"
})
```

```python
data.columns
```

```python
#collect columns which are no use
cols=["ID","Year_Birth","Marital_Status","Teenhome","Kidhome","Dt_Customer"]
spending_cols=["MntWines", "MntFruits",
       "MntMeatProducts", "MntFishProducts", "MntSweetProducts",
       "MntGoldProds"]
cols_to_drop=cols+spending_cols
data_cleaned=data.drop(columns=cols_to_drop)
```

```python
data_cleaned=data_cleaned.drop_duplicates()
```

```python
num=data_cleaned.select_dtypes(include=['object']).columns
num
```

```python
#Encoding
from sklearn.preprocessing import OneHotEncoder
encode=OneHotEncoder()#in clustering case no use of drop_first=True
enc_cols=encode.fit_transform(data_cleaned[num])
```

```python
enc_data=pd.DataFrame(enc_cols.toarray(),columns=encode.get_feature_names_out(num),index=data_cleaned.index)
```

```python
#now concatinate data_cleaned and encoded data
data_encoded=pd.concat([data_cleaned.drop(columns=num),enc_data],axis=1)
```

```python
data_encoded.shape
```

```python
X=data_encoded
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
#visualized
from sklearn.decomposition import PCA
pca=PCA(n_components=3) #two components pc1 and pc2
X_pca=pca.fit_transform(X_scaled)
```

```python
fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111,projection="3d")
plt.scatter(X_pca[:,0],X_pca[:,1],X_pca[:,2])
ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
ax.set_zlabel("PCA3")
ax.set_title("PCA")
```

```python
a=pca.explained_variance_ratio_
print(a)
```

```python
sum=0
count=0
for x in a:
    if(sum<95):
        sum=sum+x
        count=count+1
print(sum)
print(count)
    
```

```python
#Analysis K-value:
# 1.elbow
# 2.
```

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
wcss=[]
for k in range(1,11):
    Kmeans=KMeans(n_clusters=k,random_state=42)
    Kmeans.fit_predict(X_pca)
    wcss.append(Kmeans.inertia_)
```

```python
knee=KneeLocator(range(1,11),wcss,curve="convex",direction="decreasing")
optimal_k=knee.elbow
```

```python
print("best k",optimal_k)
```

```python
from sklearn.cluster import KMeans
Kmeans=KMeans(n_clusters=4,random_state=42)
y_predict=Kmeans.fit_predict(X_pca)
```

```python
#silhouette score
from sklearn.metrics import silhouette_score
scores=[]
for k in range(2,11):
    Kmeans=KMeans(n_clusters=k,random_state=42)
    labels=Kmeans.fit_predict(X_pca)
    score=silhouette_score(X_pca,labels)
    scores.append(score)

plt.plot(range(2,11),scores,marker='o')
plt.xlabel("k")
plt.ylabel("sh")
```

```python
k_range=range(2,11)
fig,ax1=plt.subplots(figsize=(8,6))
ax1.plot(k_range,wcss[:len(k_range)],marker="o",color="blue")
ax1.set_xlabel("k")
ax1.set_ylabel("wcss")

ax2=ax1.twinx()
ax2.plot(k_range,scores[:len(k_range)],marker="x",color="red",linestyle="--")
ax2.set_ylabel("SS")
```

```python
fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111,projection="3d")
plt.scatter(X_pca[:,0],X_pca[:,1],X_pca[:,2],c=y_predict)
```

```python
#agglomerative clustering
from sklearn.cluster import AgglomerativeClustering
agg_clf=AgglomerativeClustering(n_clusters=4,linkage="ward")
labels=agg_clf.fit_predict(X_pca)
fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111,projection="3d")
plt.scatter(X_pca[:,0],X_pca[:,1],X_pca[:,2],c=labels)
```
